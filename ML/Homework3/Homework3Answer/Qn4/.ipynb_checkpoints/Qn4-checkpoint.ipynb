## Stochastic Gradient decent modified from question 1,
## It stores value of min risk which will be replace everytime lower risk is calculated
## n = no of iterations, x and y are features and labels, tetar is the weight. 
## It returns the minimum risk from all the n iterations, 
## tetarList = weights of Minimum risk which will also be replaced
## points = is a list that stores all the risk every 100 iterations
## Weights = is a list of the theta every 100 iterations

```python
#Modified From Hmw1
def StochasticGradientDecent(n,lrate,x,y,tetar):
    minRisk=logRegressRisk(x,y,tetar)
    tetarList=tetar
    points = []
    weights = []
    for j in range(1,n+1):
        i = random.randint(0,len(x)-1)
        b = (-y[i]*x[i])/(1+math.exp(y[i]*np.matmul(x[i],tetar)))
        tetar = np.subtract(tetar,lrate*np.transpose(b))
        if (logRegressRisk(x,y,tetar)<minRisk):
            tetarList=tetar
            minRisk = logRegressRisk(x,y,tetar)
        if (j%100 == 0):
            points.append(minRisk)
            weights.append(tetarList)
        
    return(minRisk,tetarList,points,weights)
```

## This is a function to calculate the empirical risk of the logistic Regression

```python
def logRegressRisk(x,y,tet):
    risk = 0
    a = -1*np.array(y) * np.array(np.matmul(x,tet))
    for i in a:
        risk += math.log(1+math.exp(i))
    return(risk/len(y))
```

```python
ans = StochasticGradientDecent(10000,0.1,x,y,tet)
ans
```

```python
ans[3]
```

```python
Y = []
for tetar in ans[3]:
    log = 0
    a =-1*np.array(y) * np.array(np.matmul(x,tetar))
    for i in a:
        log += math.log(1/(1+math.exp(i)))
    Y.append(log)
Y
len(Y)
```

```python
import matplotlib.pyplot as plt
plt.plot(Y)
plt.ylabel('Log-likelihood value')
plt.xlabel('No of Iterations in hundreds')
```
