```python
def emission(x,y,df):
    #uniqueLabels= df[1].unique()
    countY = len(df.loc[df[1]==y])
    countYtoX = len(df.loc[(df[0]==x)&(df[1]==y)])
    return(countYtoX/countY)
```

```python
def smoothEmission(k,dfTrain,dfTest):
    xTrain = dfTrain[0]
    yTrain = dfTrain[1]
    xTest = dfTest[0]
    uniqueWordsTrain = xTrain.unique()
    uniqueWordsTest = xTest.unique()
    for i in uniqueWordsTrain: 
        if (len(df.loc[df[0]==i])<k):
            #replace i with unk return new df
            xTrain = xTrain.replace(i,"unk")
    dfTrain[0] = xTrain
    uniqueWordsTrain = xTrain.unique()
    
    for j in uniqueWordsTest:
        if (j not in uniqueWordsTrain):
            #replace j with unk return new df
            xTest = xTest.replace(j,"unk")                
    dfTest[0] = xTest
    
    return(dfTrain,dfTest)
```

```python
def labelGenerator(dfTrain,dfTest):
    newDF = smoothEmission(3,dfTrain,dfTest)
    dfTrain,dfTest = newDF[0],newDF[1]
    #e = emission(dfTrain[0],dfTrain[1],newDF[0])
    d ={}
    maxVal={}
    w={}
    for i in range(len(dfTrain)):
        if (dfTrain[0][i] not in maxVal):
            maxVal[dfTrain[0][i]] = emission(dfTrain[0][i],dfTrain[1][i],dfTrain)
            d[dfTrain[0][i]] = dfTrain[1][i]
            w[dfTrain[0][i]] = [dfTrain[1][i]]
        elif(dfTrain[1][i] not in w[dfTrain[0][i]] ):
            w[dfTrain[0][i]].append(dfTrain[1][i])
            if(emission(dfTrain[0][i],dfTrain[1][i],dfTrain) > maxVal[dfTrain[0][i]]):
                maxVal[dfTrain[0][i]] = emission(dfTrain[0][i],dfTrain[1][i],dfTrain)
                d[dfTrain[0][i]] = dfTrain[1][i]
        print("row"+str(i))
    return(d)
    
```

```python
predicted = labelGenerator(df,dfTest)
predicted
```
