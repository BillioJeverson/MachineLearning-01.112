```python
import copy
newLs = copy.deepcopy(ls)
ls.reverse()
newLs= newLs + ls
```

```python
f = open(r"C:\Users\user\Documents\MachineLearning01.112\MachineLearning-01.112\ML\Project\EN\trainNew","w", encoding="utf-8")

for i in range (len(newLs)):
    f.write(newLs[i])
f.close()
```
